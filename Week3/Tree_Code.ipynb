```python
import pandas as pd
```

```python
spam = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data",header = None)
```

```python
spam.columns = ["Make", "Address", "All" , "3d", "Our", "Over", "Remove","Internet",
               "Order", "Mail","Receive","Will","People","Report","Addresses",
               "Free","Business","Email","You","Credit","Your","Font","000",
               "Money","Hp","Hpl","George","650","lab","labs","telnet","857",
               "data","415","85","technology","1999","parts","pm","Direct",
               "cs","meeting","original","project","re","edu","table","conference",
               "freq_;","freq_(","freq_[","freq_!","freq_$","freq_#",
               "capital_run_length_average","Capital_run_length_longest",
               "capital_run_length_total","Spam"]
```

```python
spam
```

```python
def f(row):
    if row['Spam'] == 1:
        val = 'Spam'
    else:
        val = 'Email'
    return val

spam['Spam_Cat'] = spam.apply(f,axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(spam, test_size=.25, random_state=123)
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn import tree
```

```python
X_train = train.iloc[:, 0:56]
```

```python
Y_train = train.iloc[:,58]
Y_train
```

```python
from sklearn import tree
help(tree.DecisionTreeClassifier)
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy",min_impurity_decrease = .01)
clf.fit(X_train,Y_train)
```

```python
X_test = test.iloc[:,0:56]
Y_test = test.iloc[:,58]
```

```python
y_test_fit = clf.predict(X_test)
```

```python
import sklearn.metrics
```

```python
confusion = sklearn.metrics.confusion_matrix(Y_test,y_test_fit)
```

```python
confusion/ confusion.sum(axis=1)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0).fit(X_train, Y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
confusion = sklearn.metrics.confusion_matrix(Y_test,y_pred)
confusion/confusion.sum(axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier().fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)

confusion = sklearn.metrics.confusion_matrix(Y_test,y_pred)
confusion/confusion.sum(axis=1)
```
